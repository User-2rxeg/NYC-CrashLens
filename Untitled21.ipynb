## Handle Outliers in 'NUMBER OF PERSONS INJURED' (df_crashes)

### Subtask:
Cap extreme values in the 'NUMBER OF PERSONS INJURED' column of `df_crashes` using the IQR method. Values above Q3 + 1.5 * IQR will be capped at Q3 + 1.5 * IQR. This prevents extreme counts from disproportionately influencing analyses. After capping, verify the new maximum value for the column.


**Reasoning**:
To cap the extreme values in the 'NUMBER OF PERSONS INJURED' column, I will first calculate the Q1, Q3, and IQR, then determine the upper bound, apply the capping, and finally verify the new maximum value.



```python
Q1 = df_crashes['NUMBER OF PERSONS INJURED'].quantile(0.25)
Q3 = df_crashes['NUMBER OF PERSONS INJURED'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Cap values above the upper bound
initial_max = df_crashes['NUMBER OF PERSONS INJURED'].max()
df_crashes['NUMBER OF PERSONS INJURED'] = np.where(
    df_crashes['NUMBER OF PERSONS INJURED'] > upper_bound,
    upper_bound,
    df_crashes['NUMBER OF PERSONS INJURED']
)

print(f"Original max value in 'NUMBER OF PERSONS INJURED': {initial_max}")
print(f"Calculated upper bound for capping: {upper_bound}")
print(f"New max value in 'NUMBER OF PERSONS INJURED' after capping: {df_crashes['NUMBER OF PERSONS INJURED'].max()}")
```

## Handle Outliers in 'VEHICLE_ID' (df_persons)

### Subtask:
Cap extreme values in the 'VEHICLE_ID' column of `df_persons` using the IQR method. Values above Q3 + 1.5 * IQR will be capped at Q3 + 1.5 * IQR, and values below Q1 - 1.5 * IQR will be capped at Q1 - 1.5 * IQR. After capping, verify the new minimum and maximum values for the column.

**Reasoning**:
To cap the extreme values in the 'VEHICLE_ID' column, I will first calculate Q1, Q3, and IQR, then determine the upper and lower bounds, apply the capping using `np.where`, and finally verify the new minimum and maximum values. This will ensure that `VEHICLE_ID` values fall within a reasonable range.

```python
Q1_vehicle_id = df_persons['VEHICLE_ID'].quantile(0.25)
Q3_vehicle_id = df_persons['VEHICLE_ID'].quantile(0.75)
IQR_vehicle_id = Q3_vehicle_id - Q1_vehicle_id

upper_bound_vehicle_id = Q3_vehicle_id + 1.5 * IQR_vehicle_id
lower_bound_vehicle_id = Q1_vehicle_id - 1.5 * IQR_vehicle_id

# Store initial min/max for verification
initial_min_vehicle_id = df_persons['VEHICLE_ID'].min()
initial_max_vehicle_id = df_persons['VEHICLE_ID'].max()

# Cap values
df_persons['VEHICLE_ID'] = np.where(
    df_persons['VEHICLE_ID'] > upper_bound_vehicle_id,
    upper_bound_vehicle_id,
    np.where(
        df_persons['VEHICLE_ID'] < lower_bound_vehicle_id,
        lower_bound_vehicle_id,
        df_persons['VEHICLE_ID']
    )
)

print(f"Original min value in 'VEHICLE_ID': {initial_min_vehicle_id}")
print(f"Original max value in 'VEHICLE_ID': {initial_max_vehicle_id}")
print(f"Calculated lower bound for capping: {lower_bound_vehicle_id}")
print(f"Calculated upper bound for capping: {upper_bound_vehicle_id}")
print(f"New min value in 'VEHICLE_ID' after capping: {df_persons['VEHICLE_ID'].min()}")
print(f"New max value in 'VEHICLE_ID' after capping: {df_persons['VEHICLE_ID'].max()}")
```
